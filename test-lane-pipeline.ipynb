```python
import matplotlib.image as mpimg
import scipy.misc as spmisc
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
default_color_space='RGB'
```

```python
# helper to read an image. adds ability to change underlying implementation of image reading function
def read_image(file):
    return spmisc.imread(file, mode=default_color_space)
```

```python
image = read_image('test_images/test5.jpg')
#image = read_image('test_images/mytest1.jpg')
plt.imshow(image);
image.shape
```

```python
from LaneDetector import *
```

```python
with open("calibration_pickle.p", mode='rb') as f:
    dist_pickle = pickle.load(f)
```

```python
detector2 = LaneDetector(dist_pickle['mtx'], dist_pickle['dist'])
```

```python
for i in range(10):
    detector2.process_image(image)

# Visualize
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20,10));
ax1.imshow(detector2._undistorted_image);
ax1.set_title('Undistorted Image', fontsize=30);
ax2.imshow(detector2._top_down_binary_image, cmap='gray');
ax2.set_title('Top down binary', fontsize=30);
ax3.imshow(detector2._top_down_binary_with_lines_image);
ax3.set_title('Annotated Top down binary', fontsize=30);
ax4.imshow(detector2._original_annotated_image);
ax4.set_title('Annotated Undistorted image', fontsize=30);
```
